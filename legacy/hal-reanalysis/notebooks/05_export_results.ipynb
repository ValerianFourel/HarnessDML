# 05 — Export results (manifest → size guard → push)

**NETWORK REQUIRED for the final push** (login/JupyterHub node).
Order is mandatory: manifest, stage `results/` (+ `schema/` in Phase 1),
size guard, then commit + push. The guard exiting non-zero means the push
is forbidden — fix, never force.

```python
from halcausal.io import manifest

run_id = manifest.new_run_id()
path = manifest.write_manifest(
    run_id,
    config_path=None,        # set to the estimand yaml for estimation runs
    seeds={},                # copy from the config actually used
    input_checksums={},      # sha256 of the decrypted slice files consumed
    counts={},               # e.g. {"rollouts": ...} — reconciliation data
)
print(run_id, "->", path)
```
